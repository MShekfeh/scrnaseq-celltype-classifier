```python
%matplotlib inline
import scanpy as sc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load your processed data from Week 1
adata = sc.read('data/processed/pbmc3k_processed.h5ad')
print(f'Loaded: {adata.n_obs} cells x {adata.n_vars} genes')
print(f'Clusters: {adata.obs.leiden.unique()}')
```

```python
# Map leiden cluster numbers to known PBMC cell types
# Based on canonical marker genes for PBMC 3k
cluster_labels = {
    '0': 'CD4 T cells',
    '1': 'CD14 Monocytes',
    '2': 'CD4 T cells',
    '3': 'B cells',
    '4': 'CD8 T cells',
    '5': 'NK cells',
}

adata.obs['cell_type'] = adata.obs['leiden'].map(cluster_labels)
print(adata.obs['cell_type'].value_counts())

# Visualize labeled UMAP
sc.pl.umap(adata, color='cell_type', title='PBMC cell types', save='_celltypes.png')
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
from xgboost import XGBClassifier
import seaborn as sns

# Use PCA coordinates as features
X = adata.obsm['X_pca']  # shape: (n_cells, 50)
y = adata.obs['cell_type'].values

# Encode string labels to integers
le = LabelEncoder()
y_encoded = le.fit_transform(y)
print(f'Features shape: {X.shape}')
print(f'Classes: {le.classes_}')
```

```python
from sklearn.model_selection import train_test_split

# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Train XGBoost
xgb = XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.1,
    random_state=42,
    eval_metric='mlogloss',
    verbosity=0
)
xgb.fit(X_train, y_train)

# Evaluate
y_pred = xgb.predict(X_test)
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
